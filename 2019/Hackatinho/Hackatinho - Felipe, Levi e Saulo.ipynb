```python
treino = pd.read_csv("../input/train.csv")
labels = treino.Category.astype("category")
treino["Category"]  = pd.Categorical(treino.Category.astype("category")).codes
treino["DayOfWeek"] = pd.Categorical(treino.DayOfWeek.astype("category")).codes

novoTreino = treino[["Category", "DayOfWeek", "X", "Y"]]


test = pd.read_csv("../input/test.csv")
test["DayOfWeek"] = pd.Categorical(test.DayOfWeek.astype("category")).codes

novoTest = test[["DayOfWeek", "X", "Y"]]

Xtrain = novoTreino.values
Xtest = novoTest.values
```

```python
# usamos RandomForeste (um comitê de árvores de decisão)
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100, max_depth=4) # várias arvorezinhas
clf = clf.fit(Xtrain[:,1:], Xtrain[:,0].reshape(-1,1)) # o rótulo é a primeira coluna

Y = clf.predict_proba(Xtest)
```

```python
# Gerar o resultado de acordo a submissão
resultado = pd.DataFrame(Y)
resultado.index.name = "Id"
resultado.columns = set(labels)
resultado.to_csv("saida.csv")
```
